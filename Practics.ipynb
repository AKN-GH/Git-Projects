**Задание 18**

```python
def sos_until_zero():
    tot = 0
    sqtot = 0
    num = int(input('Введите целое число: '))
    tot += num
    sqtot += num ** 2
    while tot != 0:
        num = int(input('Введите целое число: '))
        tot += num
        sqtot += num ** 2
    return sqtot
sos_until_zero()
```
